```python
tips['time'].value_counts()
```

```python
print(tips[tips['time']=='Lunch']['tip'].sum())
print(tips[tips['time']=='Dinner']['tip'].sum())
```

we cannot infer from this we need to know the mean

```python
print('Lunch Mean', tips[tips['time']=='Lunch']['tip'].mean())
print('Dinner Mean',tips[tips['time']=='Dinner']['tip'].mean())
print('Lunch Median', tips[tips['time']=='Lunch']['tip'].median())
print('Dinner Median',tips[tips['time']=='Dinner']['tip'].median())
```

Looking at Mean and Median we can infer that Dinner tip is more consistent than lunch tip
